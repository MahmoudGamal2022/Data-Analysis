```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import smtplib
import pandas as pd
import json
```

```python
pages ="https://www.ida2at.com/author/aktowfik/"
heads = {"User-Agent": 
         "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36",
         "Accept-Encoding":"gzi"}
```

```python
soup = requests.get(pages, headers = heads)
soup
```

```python
soup2 = BeautifulSoup(soup.text, "html.parser")
soup2
```

```python
urls_list = []
for links in soup2.find_all("div", attrs={"class": "card-body"}):
    for link in links.find_all('a',href = True):
        urls_list.append(link['href'])
```

```python
links_list= urls_list[1: :2]
for i in links_list:
    print(i)
```

```python
pages_list = []
for links in soup2.find_all("div", attrs={"class": "page-item"}):
    for link in links.find_all('a',href = True):
        pages_list.append(link['href'])
```

```python
url_list = []
for i in pages_list:
    x= i.split("/")
    url_list.append(x)
start = url_list[0][-2]
end = url_list[-2][-2]
```

```python
page_links = []
for url in range(int(start)-1,int(end)+1):
    li ="https://www.ida2at.com/author/aktowfik/page/"+str(url)+"/"
    page_links.append(li)
```

```python
for k in page_links:
    print(k)
```

```python
for ul in page_links:
    req = requests.get(ul,headers = heads)
    soup22 = BeautifulSoup(req.text, "html.parser")
    urls_list = []
    for links in soup22.find_all("div", attrs={"class": "card-body"}):
        for link in links.find_all('a',href = True):
            urls_list.append(link['href'])
    links_list= urls_list[1: :2]
    for lin in links_list:
        re = requests.get(lin,headers = heads)
        soup23 = BeautifulSoup(re.text, "html.parser")
        p = soup23.find('h1', class_ = "post-title").text
        pr = soup23.find('main', class_ = "post-content")
        name = p+".txt"
        text_file = open(name, "w",encoding="utf-16")
        text_file.writelines(p+"\n")
        for j in pr:
            m = j.text
            text_file.writelines(str(m)+"\n")
        text_file.close()
        
```
